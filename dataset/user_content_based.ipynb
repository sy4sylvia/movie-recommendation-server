```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
# Gets the first N rows of a dataframe. N's default is 5.
movies_df.head(10)
```

```python
ratings_df.head(10)
```

```python
ratings_df = ratings_df.drop(columns=['timestamp'])
ratings_df.head(10)
```

```python
# Data processing
# get_year extracts year from the title column
def get_year(title):
    year = title[len(title) - 5:len(title) - 1]
    return year

# get_title extracts the movie title from the title column
def get_title(title):
    return title[: len(title) - 7]

movies_df.rename(columns = {'title': 'movie_year'}, inplace = True)

movies_df['title'] = movies_df['movie_year'].apply(get_title)
movies_df['year'] = movies_df['movie_year'].apply(get_year)

movies_df = movies_df.drop(columns=['movie_year'])
movies_df = movies_df[['movieId', 'title', 'genres', 'year']]
```

```python
movies_df.head(10)
```

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head(10)
```

```python
movies_with_genres_df = movies_df.copy()

# For every row in the dataframe, iterate through the list of genres and place a 1 if the movie belongs to the genre
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movies_with_genres_df.at[index, genre] = 1
        
# Fill in the NaN with 0 if the movie does not belong to the genre
movies_with_genres_df = movies_with_genres_df.fillna(0)
movies_with_genres_df.head()
```

```python
user_input = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
input_movies = pd.DataFrame(user_input)
input_movies
```

```python
# Extracting the movie ids
input_id = movies_df[movies_df['title'].isin(input_movies['title'].tolist())]
input_movies = pd.merge(input_id, input_movies)
input_movies
```

```python
input_movies = input_movies.drop(columns =['genres', 'year'])
input_movies
```

```python
user_movies = movies_with_genres_df[movies_with_genres_df['movieId'].isin(input_movies['movieId'].tolist())]
user_movies
```

```python
user_genres = user_movies.drop(columns = ['movieId', 'title', 'genres', 'year'])
user_genres
```

```python
input_movies['rating']
```

```python
transpose_genres = user_genres.transpose()
transpose_genres.shape
```

```python
# user_profile = transpose_genres.dot(input_movies)
user_profile = np.matmul(transpose_genres, input_movies['rating'])
user_profile
```

```python
movies_with_genres_df
```

```python
# Similarly, get the genres of every movie in the original dataframe
genres = movies_with_genres_df.set_index(movies_with_genres_df['movieId'])
genres.head(10)
```

```python
genres = genres.drop(columns = ['movieId', 'title', 'genres', 'year'])
genres.head(10)
```

```python
genres.shape
```

```python
# Multiply the genres by the weights and take the weighted average
recommendation_df = ((genres * user_profile).sum(axis = 1)) / (user_profile.sum())
recommendation_df.head()
```

```python
# Recommendation System, sort the weighted average descendingly
recommendation_df = recommendation_df.sort_values(ascending = False)
recommendation_df.head(10)
```

```python
# The recommended movies with additional info, return the top 10 movies
top_10_reco = movies_df.loc[movies_df['movieId'].isin(recommendation_df.head(10).keys())]
top_10_reco
```
